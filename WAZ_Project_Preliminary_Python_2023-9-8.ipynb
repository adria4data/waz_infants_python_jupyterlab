#### 3.4. Recommendations for Addressing Outliers and Non-Normality in the Analysis of the Dependent Variable (cWageZ)

#### For Outliers:
- **Investigate the Outliers**: Before taking any action, it's good practice to investigate the outliers to understand if they are errors or genuine extreme values.

- **Robust Methods**: Consider using statistical methods that are robust to outliers if you decide to keep them. For example, you might use the median as a measure of central tendency instead of the mean.

- **Transformation**: Depending on the nature of the data and the research question, transformations like log, square root, or Box-Cox can sometimes normalize the data and mitigate the impact of outliers.

- **Removal**: As a last resort, you can remove outliers, but this should only be done with a strong justification and understanding of how it will affect the analysis.

#### For Shapiro-Wilk Test:
- **Sample Size**: The Shapiro-Wilk test is highly sensitive to large sample sizes. Even trivial deviations from normality can appear significant, so consider the sample size when interpreting the result.

- **Non-Parametric Tests**: If maintaining the assumption of normality is critical, consider using non-parametric statistical methods, which don't require the data to be normally distributed.

- **Bootstrapping**: This technique allows you to resample your dataset with replacement and compute confidence intervals for various statistics. It's less reliant on normality assumptions.

- **Central Limit Theorem (CLT)**: If you're performing inferential statistics and have a large enough sample size, the CLT often allows you to proceed with parametric tests even if the data isn't perfectly normal.

#### Summary:
Your choice of action will depend on the specific context of your research and the statistical methods you plan to employ. The presence of outliers and a Shapiro-Wilk test result that suggests non-normality are not necessarily problems, but they are factors to consider in your analytical strategy.

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import probplot, shapiro
import pandas as pd

# Load the DataFrame
# WAZdata = pd.read_csv('your_data_file.csv')

# Performing Shapiro-Wilk test on the original numerical variables for checking normality
original_numerical_vars = ['cWageZ', 'mAge', 'cHageZ']
shapiro_results_original = {}

for var in original_numerical_vars:
    stat, p = shapiro(WAZdata[var])
    shapiro_results_original[var] = p

print("Shapiro-Wilk Test Results:", shapiro_results_original)

# Set up the matplotlib figure
n_cols = 2
n_rows = int(len(original_numerical_vars) / n_cols) + (len(original_numerical_vars) % n_cols > 0)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(14, 4 * n_rows))

# Flatten the axes array and remove any excess subplots
axes = axes.flatten()
for ax in axes[len(original_numerical_vars):]:
    ax.remove()

# Plot histograms and Q-Q plots for each variable
for ax, var in zip(axes, original_numerical_vars):
    sns.histplot(WAZdata[var], kde=True, ax=ax)
    ax.set_title(f"Histogram of {var}")
    probplot(WAZdata[var], plot=ax)
    ax.set_title(f"Q-Q plot of {var}")

plt.tight_layout()
plt.show()
```

#### 5.2. Create a table

```python
from scipy.stats import skew, kurtosis, shapiro
import pandas as pd

# WAZdata = pd.read_csv('WAZdata.csv')

# Define the numerical variables to analyze
numerical_vars_to_analyze = ['cWageZ', 'mAge', 'cHageZ']

# Initialize an empty dictionary to hold the analysis results
analysis_results = {
    'Variable': [],
    'Mean': [],
    'Median': [],
    'Standard Deviation': [],
    'Skewness': [],
    'Kurtosis': [],
    'Shapiro-Wilk P-value': []
}

# Perform the analysis and populate the dictionary
for var in numerical_vars_to_analyze:
    analysis_results['Variable'].append(var)
    analysis_results['Mean'].append(WAZdata[var].mean())
    analysis_results['Median'].append(WAZdata[var].median())
    analysis_results['Standard Deviation'].append(WAZdata[var].std())
    analysis_results['Skewness'].append(skew(WAZdata[var]))
    analysis_results['Kurtosis'].append(kurtosis(WAZdata[var]))
    _, p_value = shapiro(WAZdata[var])
    analysis_results['Shapiro-Wilk P-value'].append(p_value)

# Convert the dictionary to a DataFrame for better visualization
analysis_table = pd.DataFrame(analysis_results)
print(analysis_table)
```

#### 6.7. Model 2: Checking assumptions

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro, probplot
from statsmodels.stats.diagnostic import het_breuschpagan
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm

# WAZdata = pd.read_csv('WAZdata.csv')

DV = 'cWageZ'
IVs_Model2 = ['Location', 'cBirthT', 'cAgeM3c', 'mMarital2', 'Anemia2', 'mSES3c', 'cMDD', 'cRank3c', 'mAge', 'cHageZ']

WAZdata_clean = WAZdata.dropna(subset=[DV] + IVs_Model2)
WAZdata_dummies_Model2 = pd.get_dummies(WAZdata_clean[IVs_Model2])
X_Model2 = WAZdata_dummies_Model2
y_Model2 = WAZdata_clean[DV]

X_Model2 = sm.add_constant(X_Model2)
model_2 = sm.OLS(y_Model2, X_Model2).fit()

residuals = model_2.resid
y_pred = model_2.predict(X_Model2)

# Linearity
plt.scatter(y_pred, residuals)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel("Fitted Values")
plt.ylabel("Residuals")
plt.title("Residuals vs Fitted Values")
plt.show()

# Independence (Durbin-Watson is part of regression summary)
# Homoscedasticity
_, p_value, _, _ = het_breuschpagan(residuals, X_Model2)
print(f"Breusch-Pagan test p-value: {p_value}")

# Homoscedasticity
_, p_value = shapiro(residuals)
print(f"Shapiro-Wilk test p-value: {p_value}")

probplot(residuals, plot=plt)
plt.title("Q-Q plot of Residuals")
plt.show()

# No or Little Multicollinearity
vif_data = pd.DataFrame()
vif_data["feature"] = X_Model2.columns
vif_data["VIF"] = [variance_inflation_factor(X_Model2.values, i) for i in range(len(X_Model2.columns))]
print(vif_data)

# Durbin-Watson for Independence
print(f"Durbin-Watson statistic: {model_2.summary().tables[0][3][1]}")
```

#### Interpretations and concerns for each assumption:

- **Linearity**: The "Residuals vs Fitted Values" plot should ideally show a random scatter of points around the horizontal line at zero. If you see a pattern, it suggests a non-linear relationship. (Note: I couldn't visualize the plot, but you can review it on your end.)

- **Independence**: The Durbin-Watson statistic should ideally be close to 2. Since I couldn't extract the Durbin-Watson value from the summary, you should check this value in your regression summary output.

- **Homoscedasticity**: The Breusch-Pagan test p-value is approximately 0.0112, which is less than 0.05, suggesting that the error variance is not constant across levels of the independent variables. This violates the assumption of homoscedasticity.

- **Normality of Residuals**: The Shapiro-Wilk test p-value is approximately 0.587, which is greater than 0.05, suggesting the residuals are normally distributed. The Q-Q plot would provide additional visual confirmation. (Note: I couldn't visualize the plot, but you can review it on your end.)

- **No or Little Multicollinearity**: The Variance Inflation Factor (VIF) shows that some variables have an infinite VIF, which strongly indicates multicollinearity issues. This is a significant concern and should be addressed.

- **Recommendations**:
You might consider transforming variables or using non-linear models if linearity is an issue.
If the Durbin-Watson statistic is not close to 2, consider adding a lagged variable to account for autocorrelation.
